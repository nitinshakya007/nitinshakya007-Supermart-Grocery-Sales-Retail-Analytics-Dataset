```python
# Step 1: Data Loading and Preparation
data = pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset.csv')
data.head(2)
```

```python
# Convert 'Order Date' to datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d-%m-%Y', errors='coerce')
```

```python
# Extract year and month from 'Order Date'
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
```

## Step 2: Data Analysis

```python
# Sales Distribution by Category
sales_by_category = data.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("Sales by Category:\n", sales_by_category)
```

```python
# Yearly Sales Trend
yearly_sales = data.groupby('Year')['Sales'].sum().reset_index()
print("Yearly Sales:\n", yearly_sales)
```

```python
# Monthly Sales Trend
monthly_sales = data.groupby('Month')['Sales'].sum().reset_index()
print("Monthly Sales:\n", monthly_sales)
```

```python
# Profit vs Discount Analysis
profit_discount_correlation = data[['Profit', 'Discount']].corr().iloc[0, 1]
print(f"Correlation between Profit and Discount: {profit_discount_correlation}")
```

## Step 3: Data Visualization

```python
# Sales Distribution by Category
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_by_category.index, y=sales_by_category.values, palette='viridis')
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Yearly Sales Trend
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Sales', data=yearly_sales, marker='o', color='b')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Monthly Sales Trend
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Sales', data=monthly_sales, marker='o', color='g')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
# Profit vs Discount Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', data=data, hue='Category', palette='Set1')
plt.title('Profit vs Discount by Category')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.legend(loc='upper right')
plt.show()
```

## Step 4: Feature Engineering and Model Building

```python
# Encode categorical variables
label_encoder = LabelEncoder()
data['Category'] = label_encoder.fit_transform(data['Category'])
data['Sub Category'] = label_encoder.fit_transform(data['Sub Category'])
data['City'] = label_encoder.fit_transform(data['City'])
data['Region'] = label_encoder.fit_transform(data['Region'])
data['State'] = label_encoder.fit_transform(data['State'])
```

```python
# Feature Selection
features = data[['Category', 'Sub Category', 'City', 'Region', 'State', 'Discount', 'Month', 'Year']]
target = data['Sales']
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Model Building - Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Model Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
